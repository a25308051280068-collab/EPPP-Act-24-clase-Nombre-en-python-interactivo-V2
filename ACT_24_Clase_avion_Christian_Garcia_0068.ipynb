```python
class Avion:
    def __init__(self, modelo, aerolinea, capacidadPasajeros, autonomiaVuelo):
        self.modelo = modelo
        self.aerolinea = aerolinea
        self.capacidadPasajeros = capacidadPasajeros
        self.autonomiaVuelo = autonomiaVuelo

    def despegar(self):
        print(f"El avión modelo {self.modelo} de {self.aerolinea} está despegando.")

    def aterrizar(self):
        print(f"El avión modelo {self.modelo} ha aterrizado con éxito.")

    def anunciarDestino(self):
        print(f"Anunciando destino para el vuelo en el avión {self.modelo}.")

# --- Lógica con Input para crear los objetos (ejemplo) ---

print("--- REGISTRO DE AVIONES ---")

# Datos para el Avion 1
print("\nIngrese los datos para el Avion 1:")
modelo1 = input("Modelo: ")
aerolinea1 = input("Aerolínea: ")
capacidad1 = input("Capacidad de Pasajeros: ")
autonomia1 = input("Autonomía de Vuelo (km): ")
avion1 = Avion(modelo1, aerolinea1, capacidad1, autonomia1)

# Datos para el Avion 2
print("\nIngrese los datos para el Avion 2:")
modelo2 = input("Modelo: ")
aerolinea2 = input("Aerolínea: ")
capacidad2 = input("Capacidad de Pasajeros: ")
autonomia2 = input("Autonomía de Vuelo (km): ")
avion2 = Avion(modelo2, aerolinea2, capacidad2, autonomia2)

# --- Ejecución de las funciones ---

print("\n" + "="*30)
print(f"--- Actividades de {avion1.modelo} ({avion1.aerolinea}) ---")
avion1.despegar()
avion1.anunciarDestino()
avion1.aterrizar()

print("\n" + "="*30)
print(f"--- Actividades de {avion2.modelo} ({avion2.aerolinea}) ---")
avion2.despegar()
avion2.aterrizar()
avion2.anunciarDestino()

print("\n" + "="*30)
print("Programa terminado")
print("Programa realizado por Christian Garcia NL = 21")
```
